```python
df = pd.read_csv("../judgements/judgements_writing_prompts_HQ_20250312_113801.csv")

# Filter to just a single judge, so it's the right shape
df = df[df["judge_model"] == "google/gemini-2.0-flash-001"]
# Filter to just the first 20 problems
df = df[df["question_id"] < 20]

# Sort by question_id, response_id, criteria
df = df.sort_values(["question_id", "response_id", "criteria"])

print(f"Size of df is {len(df)}")
print(f"Number of distinct questions: {len(df['question_id'].unique())}")
print("Number of unique response models: ", len(df["response_model"].unique()))
print(f"Number of judgements per question: {len(df) / len(df['question_id'].unique())}")


# Replace the judgement_ids column with new judgement_ids using str(uuid.uuid4())[:13]
df["judgement_id"] = [str(uuid.uuid4())[:13] for _ in range(len(df))]

# Replace the judgement_criteria_id column with new judgement_criteria_ids using str(uuid.uuid4())[:13]
df["judgement_criteria_id"] = [str(uuid.uuid4())[:13] for _ in range(len(df))]

# Replace the judge_model column with "human" 
df["judge_model"] = "human"

# Replace the judge_reasoning column with "N/A"
df["judge_reasoning"] = "N/A"

# Replace the judge_rating with 0
df["judge_rating"] = 0

df.head(80)
# df.head(60)

```
