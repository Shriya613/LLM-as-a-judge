```python
import pandas as pd
```

```python
def get_output_filename(input_file):
    """Creates output filename by inserting '_corrected' before the file extension"""
    base, ext = input_file.rsplit('.', 1)
    return f"{base}_corrected.{ext}"
```

```python
input_file = "HumanAnnotationCreativeWriting.csv"
# input_file = "human_judgements_writing_prompts.csv"
output_file = get_output_filename(input_file)
print(f"Input file: {input_file}")
print(f"Output file: {output_file}\n\n")

df = pd.read_csv(input_file)

print(f"Before correction, we have {len(df)} rows \n")
print(f"Criteria are: {df['criteria'].unique()}")

# Count occurrences of each judgement_id before correction
judgement_id_counts = df['judgement_id'].value_counts()
print(judgement_id_counts[:10])



```

```python
def create_judgement_id_from_response_id(response_id):
    """
    Transforms a response_id into a judgement_id by incrementing 
    each numeric character by 1 (wrapping 9 to 0).
    """
    result = ""
    for char in response_id:
        if char.isdigit():
            # Increment and wrap 9 to 0
            new_digit = (int(char) + 1) % 10
            result += str(new_digit)
        else:
            # Keep non-numeric characters as they are
            result += char
    return result

test_input = "23a1bb3c-2ce8"
print(create_judgement_id_from_response_id(test_input))
```

```python
# Assuming df is your DataFrame with human annotations
def update_judgement_ids(df):
    # Create a mapping from response_id to new judgement_id
    unique_response_ids = df['response_id'].unique()
    id_mapping = {
        resp_id: create_judgement_id_from_response_id(resp_id) 
        for resp_id in unique_response_ids
    }
    
    # Apply the mapping to create the new judgement_id column
    df['judgement_id'] = df['response_id'].map(id_mapping)
    
    return df
```

```python
new_df = update_judgement_ids(df)
```
