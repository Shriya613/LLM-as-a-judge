```python
import pandas as pd
```

```python
# cr_lm_df = pd.read_csv("../judgements/judgements_writing_prompts_HQ_20250312_113801.csv") # OG
# cr_human_df = pd.read_csv("../annotation/HumanAnnotationCreativeWriting_corrected.csv") # Annotations
cr_combo_filtered_df = pd.read_csv("../judgements/judgements_writing_prompts_combined_with_humans.csv") # Combo

# qa_lm_df = pd.read_csv("../judgements/judgements_eli5_HQ_20250311_213806.csv")
# qa_human_df = pd.read_csv("../annotation/HumanAnnotationQuestionAnswering_corrected.csv")
qa_combo_filtered_df = pd.read_csv("../judgements/judgements_eli5_combined_with_humans.csv")
```

```python
# These are responses that we identified as good during human annotation
good_response_ids = [
    "d1b6f45f-ffca",
    "7e6dcd68-bf6c",
    "5e2f5eaf-543c",
    "68875055-73b2",
    "e7b63a9f-4e5c",
    "cbddbd52-42dc",
]

bad_response_ids = [
    "69cef837-53a6",
    "affafaa6-0218",
    "8a88b1d0-d8c4",
    "e0ed521f-abb4",
    "bab69b5a-9ad0"
]
```

```python
for response_id in bad_response_ids:
    # First determine which dataset this response is from by checking both combo dataframes
    if response_id in cr_combo_filtered_df['response_id'].values:
        df = cr_combo_filtered_df
    else:
        df = qa_combo_filtered_df
        
    # Get all data for this response
    response_rows = df[df['response_id'] == response_id]
    
    # Get first row for response details
    response_details = response_rows.iloc[0]
    
    # Get unique criteria for this dataset
    criteria = response_rows['criteria'].unique()
    
    print(f"\nDataset: {'CR' if response_id in cr_combo_filtered_df['response_id'].values else 'QA'}")
    print(f"Question ID: {response_details['question_id']}")
    print(f"Question: {response_details['question']}")
    print(f"Response ID: {response_id}")
    print(f"Response model: {response_details['response_model']}")
    print(f"Response: {response_details['response']}\n")
    
    print("\nRatings by Criteria:")
    for c in criteria:
        # Get human rating
        human_rating = response_rows[
            (response_rows['criteria'] == c) & 
            (response_rows['judge_model'] == 'human')
        ]['judge_rating'].iloc[0]
        
        # Get average model rating (excluding human)
        model_rating = response_rows[
            (response_rows['criteria'] == c) & 
            (response_rows['judge_model'] != 'human')
        ]['judge_rating'].mean()
        
        print(f"{c}:")
        print(f"  Human rating: {human_rating:.2f}")
        print(f"  Model average: {model_rating:.2f}")
    
    print("-" * 80)
```
